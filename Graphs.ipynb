```python
from pandas import read_csv
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.figsize'] = [10, 30]
# load dataset
values = data[0:3649].values
# specify columns to plot
groups = np.arange(13)
i = 1
# plot each column
plt.figure()
c=['black','coral','palegreen','red','cyan','orange','magenta','yellow','pink','tan','violet','lightseagreen','salmon','olive']
for group in groups:
 plt.subplot(len(groups), 1, i)
 plt.plot(values[:, group],color=c[i])
 plt.title(data.columns[group], y=0.5, loc='center',fontweight='bold')
 i += 1
plt.title('Year 1969-1979')
plt.show()
```

```python

values = data[0:3651].values
# specify columns to plot
groups = np.arange(13)
i = 1
# plot each column
fig, axs = plt.subplots(len(groups), 1, figsize=(10, 30))
c=['coral','palegreen','red','cyan','orange','magenta','yellow','pink','tan','violet','lightseagreen','salmon','olive']
for i, group in enumerate(groups):
    axs[i].plot(values[:, group], color=c[i])
    axs[i].set_title(data.columns[group], y=0.5, loc='center', fontweight='bold')
    plt.xlabel('No. of days')
fig.suptitle('Year 1969-1979', fontsize=16, fontweight='bold',  ha='center', va='top')

plt.savefig('Plot_1.png')
plt.show()
```
